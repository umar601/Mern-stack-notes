```python
height_train=[[1.6], [1.65],[1.7],[1.73],[1.8]]
weigth_train=[60,65,70,75,80]
height_test=[[1.58],[1.62],[1.69],[1.76],[1.82]]
weight_test = [58, 63, 72, 73, 85]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(height_train,weigth_train)
```

```python
model.predict([[1.75]])
```

```python
plt.scatter(height_train,weigth_train,label="training data")
```

```python
min_height = min(height_train)
max_height = max(height_train)

x_line = [min_height,max_height]
y_line = model.predict(x_line)
```

```python
plt.plot(x_line,y_line,label="regression line" )
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight - Linear Regression')
plt.legend()
plt.show()
```

```python
model.coef_
```

```python
model.intercept_
```

```python
weight_predicted=model.predict(height_train)
weight_predicted
```

```python
rss = ((weigth_train - weight_predicted)**2).sum()
rss
```

```python
r2 = model.score(height_test, weight_test)

print(f"R-squared for test data: {r2:.4f}")

# Interpretation
if r2 > 0.7:
    print("Interpretation: Strong fit - The model explains", f"{r2*100:.1f}%", "of the variance in test data weights")
elif r2 > 0.5:
    print("Interpretation: Moderate fit - The model explains", f"{r2*100:.1f}%", "of the variance in test data weights")
else:
    print("Interpretation: Weak fit - The model explains only", f"{r2*100:.1f}%", "of the variance in test data weights")
```

```python
def detectOutliers(df, col):
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  iqr = Q3 - Q1
  lower_bound = Q1 - 1.5 * iqr
  upper_bound = Q3 + 1.5 * iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  return outliers, lower_bound, upper_bound

for col in outlierDf.select_dtypes(include=np.number).columns:
  outliers_found, lower, upper = detectOutliers(outlierDf, col)

  if not outliers_found.empty:
    print(f"Column: {col}")
    print(f"Bounds: [{lower:.2f}, {upper:.2f}]")
    print(f"Outliers found: {len(outliers_found)}")
    print(outliers_found)


```

```python
def zscore_outliers(df,col):
  standardDeviation=df[col].std();
  mean=df[col].mean()
  zscore = (df[col]-mean)/standardDeviation;
  outliers = df[(zscore<-3) | (zscore>3)]
  return outliers
for col in outlierDf.select_dtypes(include=np.number).columns:
  outliers_found = zscore_outliers(outlierDf,col)
  if not outliers_found.empty:
    print(f"Column: {col}")
    print(f"Outliers found: {len(outliers_found)}")
    print(outliers_found)
```
